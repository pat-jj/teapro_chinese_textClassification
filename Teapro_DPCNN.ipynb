```python
import tensorflow as tf
print(tf.__version__)
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
PATH = './teapro.csv'
df = pd.read_csv(PATH)
df
```

```python
#删除最后一行
df.drop(index = 13948,inplace = True)

# 将float转为str
for i in range(13947):
    if(type(df.iloc[i, 2]) == float):
        df.iloc[i, 2] = ' '

# 将不干净的非零label置1, 非数字类型的置0
for i in tqdm(range(13947)):
    for j in range(7):
#         if(type(df.iloc[i, j+4]) != float):
#             df.iloc[i, j+4] = 0
        if(df.iloc[i, j+4] != 0.0):
            df.iloc[i, j+4] = 1.0

df
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test_val = train_test_split(df, test_size=0.10, random_state=101)
df_test, df_val = train_test_split(df_test_val, test_size=0.20, random_state=202)
df_train.shape, df_test.shape, df_val.shape
```

```python
print("***************============================" + "package" + "============================*****************")

df_train_ = pd.DataFrame({'content': df_train['rateContent'],'label': df_train["package"]})
df_test_ = pd.DataFrame({'content': df_test['rateContent'],'label': df_test["package"]})
df_dev_ = pd.DataFrame({'content': df_val['rateContent'],'label': df_val["package"]})

df_train_.to_csv("./DPCNN/data/train.csv", encoding="utf_8_sig", index=False)
df_test_.to_csv("./DPCNN/data/test.csv", encoding="utf_8_sig", index=False)
df_dev_.to_csv("./DPCNN/data/dev.csv", encoding="utf_8_sig", index=False)

!python ./DPCNN/run.py --train --model_name DPCNN --write_vocab True --experiment_name test
    
```
